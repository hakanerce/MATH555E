# 3. booking.com Challenge

This dataset is taken from [Booking.com Challenge ACM WSDM 2021](https://www.bookingchallenge.com/).

The original challenge was to predict the final destination city of the trip by looking at the data at hand for each user. For this question, you are free to do whatever you like. Do something interesting with the data :)

```python
booking = pd.read_csv('./data/booking/train_set.csv')
booking
```

```python
booking_n = booking.drop(['user_id' , 'city_id' , 'affiliate_id' , 'utrip_id'] , axis = 'columns')
booking_n['checkin'] =  pd.to_datetime(booking_n['checkin'])
booking_n['checkout'] =  pd.to_datetime(booking_n['checkout'])
booking_n
```

```python
booking_n.dtypes
```

```python
booking_n['timespend'] = booking_n['checkout'] - booking_n['checkin']
booking_n
```

```python
booking_n['timespend'] = booking_n['timespend'].apply(lambda x: x.value/86400000000000)
```

```python
booking_n
```

```python
le_device_class = LabelEncoder()
le_booker_country = LabelEncoder()
le_hotel_country = LabelEncoder()
```

```python
booking_n['device_class_n'] = le_device_class.fit_transform(booking_n['device_class'])
booking_n['booker_country_n'] = le_booker_country.fit_transform(booking_n['booker_country'])
booking_n['hotel_country_n'] = le_hotel_country.fit_transform(booking_n['hotel_country'])
booking_target_n = booking_n['timespend']
booking_inputs_n = booking_n.drop(['checkin' , 'checkout' , 'device_class' , 'booker_country' , 'hotel_country'] , axis = 'columns')
```

```python
booking_X_train , booking_X_test , booking_y_train , booking_y_test = train_test_split(booking_inputs_n , booking_target_n , test_size = 0.2)
```

```python
model20 = tree.DecisionTreeRegressor()
model20.fit(booking_X_train ,booking_y_train)
model20.score(booking_X_test ,booking_y_test)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

regr = linear_model.LinearRegression()

regr.fit(booking_X_train, booking_y_train)

booking_y_pred = regr.predict(booking_X_test)
booking_y_pred
```

```python
mean_squared_error(booking_y_test, booking_y_pred)
```
